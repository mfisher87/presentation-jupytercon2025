```python
from jupytergis import GISDocument

doc = GISDocument()

doc.add_raster_layer(
    name="Basemap",
    url="https://tile.openstreetmap.org/{z}/{x}/{y}.png")

doc.add_geojson_layer(
    name="France",
    path="./france_regions.geojson",
    color_expr={"fill-color": "#FF0000"},
)

doc
```
